# 2. KoBART tokenizer sample
- Tokenizer 예시입니다.
- KoBART github tokenizer 관련 부분: https://github.com/SKT-AI/KoBART#tokenizer
- Huggingface transformer tokenizer 공식 문서: https://huggingface.co/docs/transformers/main_classes/tokenizer

```python
from transformers import PreTrainedTokenizerFast

model_pretrained = 'gogamza/kobart-summarization'
tokenizer = PreTrainedTokenizerFast.from_pretrained(model_pretrained)
```

```python
sample_text = "안녕하세요 😀 삼성 SDS 자연어처리 프로젝트입니다. ^^ 태스크는 회의 녹취록 생성 요약입니다 ~~~ :)"

tokenized = tokenizer.tokenize(sample_text)
print(tokenized)
```

```python
# Convert tokens to ids
print(tokenizer.convert_tokens_to_ids(tokenized))
```

```python
# encode = tokenize + convert_tokens_to_ids
print(tokenizer.encode(sample_text))
```
